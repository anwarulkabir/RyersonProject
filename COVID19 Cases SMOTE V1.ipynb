```
import pandas as pd
import numpy as np

df = pd.read_csv ('COVID19 cases Modified V6.csv')
print(df.dtypes)
```

Convert data types fron Numeric to Categorical

```
#df = pd.read_csv ('COVID19 cases Modified V6.csv')
for col in df.columns:
  df[col] = df[col].astype(str)
print(df.dtypes)
```

```
from imblearn.over_sampling import SMOTENC
smote_nc = SMOTENC(categorical_features=[0,1,2,3,4,5,6], random_state=0)
X_resampled, y_resampled = smote_nc.fit_resample(df, df['Outcome'])
df_s = pd.DataFrame(X_resampled)


for col in df_s.columns:
  df_s[col] = df_s[col].astype(str)
print(df.dtypes)
```

Export output file to csv

```
df_s.to_csv('Final_COVID_DataSet.csv', index=False) 
files.download('Final_COVID_DataSet.csv')
df_s.info()
```
